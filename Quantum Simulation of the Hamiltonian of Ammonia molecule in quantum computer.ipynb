```python
#Using different qubit encodings of OpenFermion
import tequila as tq
import numpy
geomstring = "N 0.0 0.0 0.0\nH 0.0 0.0 1.5\nH 0.0 0.0 -1.5\nH 1.5 0.0 0.0"
basis_set = "cc-pVDZ"
active_orbitals = {"A1":[1,2,3], "A2":[1]}

# Jordan-Wigner (this is the default)
mol = tq.chemistry.Molecule(geometry=geomstring, basis_set=basis_set, active_orbitals=active_orbitals)
H = mol.make_hamiltonian()
print("Jordan-Wigner\n", H)
print("Hamiltonian has {} terms".format(len(H)))
eigenValues = numpy.linalg.eigvalsh(H.to_matrix())
print(eigenValues)
print("lowest energy = ", eigenValues[0])
```
